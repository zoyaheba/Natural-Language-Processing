```python
#importing pandas libaray to fetch the data

import pandas as pd
```

```python
df = pd.read_csv('Data.csv',encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
# diving the data into train and test
train = df[df['Date'] > '20150101']
test = df[df['Date'] > '20141231']
```

```python
#removing punctuations

data = train.iloc[:,2:27]
data.replace("[^a-zA-Z]", " ", regex = True, inplace= True)
```

```python
list1 = [i for i in range(25)]
new_Index = [str(i) for i in list1]
data.columns = new_Index
data.head()
    
```

```python
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
cv = CountVectorizer(ngram_range=(2,2))
traindata = cv.fit_transform(headlines)
```

```python
rd = RandomForestClassifier(n_estimators=200,criterion = 'entropy')
rd.fit(traindata,train['Label'])
```

```python
test_transform = []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
    
```

```python
test_data = cv.transform(test_transform)
predictions = rd.predict(test_data)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
